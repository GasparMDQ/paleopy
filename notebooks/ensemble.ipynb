# illustrates the use of the `ensemble` and `composite` classes

```python
%matplotlib inline
```

```python
import sys
from matplotlib import pyplot as plt
```

```python
jsons = './jsons/proxies/'
```

```python
sys.path.insert(0, '../')
```

```python
djsons = '../jsons/'
pjsons = '../jsons/proxies'
```

```python
from paleopy import ensemble
from paleopy import analogs
from paleopy.plotting import scalar_plot
```

```python
ens = ensemble(djsons=djsons, pjsons=pjsons, season='DJF')
```

```python
sst = analogs(ens, 'ersst', 'sst').composite()
```

```python
f = scalar_plot(sst, test=0.1).plot()
```

```python
sst.close()
```

```python
hgt = analogs(ens, 'ncep', 'hgt_1000').composite()
```

```python
f = scalar_plot(hgt, test=0.1, proj='cyl', grid=True).plot()
```
