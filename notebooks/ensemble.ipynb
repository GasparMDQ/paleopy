# illustrates the use of the `ensemble` and `composite` classes

```python
%matplotlib inline
```

```python
import sys
from matplotlib import pyplot as plt
```

```python
jsons = './jsons/proxies/'
```

```python
sys.path.append('/Users/nicolasf/CODE/paleopy/')
```

```python
from paleopy.core.ensemble import ensemble
from paleopy.core.analogs import analogs
from paleopy.plotting.scalar_plot import scalar_plot
```

```python
ens = ensemble(jsons, season='DJF')
```

```python
sst = analogs(ens, 'ersst', 'sst').composite()
```

```python
f = scalar_plot(sst, test=0.1).plot()
```

```python
sst.close()
```

```python
hgt = analogs(ens, 'ncep', 'hgt_1000').composite()
```

```python
f = scalar_plot(hgt, test=0.1, proj='cyl', grid=True).plot()
```

```python
f.savefig('/Users/nicolasf/Desktop/plot.png', dpi=200)
```
